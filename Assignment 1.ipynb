```python
import pandas as pd
toss_outcomes = ['HHH', 'HHT', 'HTH', 'HTT', 'THH', 'THT', 'TTH', 'TTT']
df3 = pd.DataFrame(toss_outcomes, columns=['toss_outcomes'])
# Count the number of outcomes with two heads and one tail
favorable_outcomes = df3[df3['toss_outcomes'].str.count('H') == 2]
Number_of_favorable_outcomes = len(favorable_outcomes)
# Calculate the probabilitty
total_outcomes = len(df3)
probability = Number_of_favorable_outcomes / total_outcomes
print('\nAnswer :')
print("Probability:", probability)
print(f"Probability: {Number_of_favorable_outcomes}/{total_outcomes}")
```

```python
import scipy.stats as stats
import numpy as np

# Specify the sample mean, standard deviation, and sample size
sample_mean = 200
sample_stddev = 30
sample_size = 2000

# Confidence levels
confidence_levels = [0.90, 0.94, 0.60]

# Calculate and display confidence intervals
for confidence_level in confidence_levels:
    alpha = 1 - confidence_level

    # Calculate the critical value (Z-score) for the given confidence level
    z_score = stats.norm.ppf(1 - alpha / 2)

    # Calculate the margin of error
    margin_of_error = z_score * (sample_stddev / np.sqrt(sample_size))

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    df11= (f"{confidence_level * 100}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
    df11 = {'C.I.':['94%','96%','98%'],
      'Lower limit':[198.738325292158,198.62230334813333,198.62230334813333],
      'Upper Limit':[201.261674707842,201.37769665186667,201.56056159570022]}
df11 =pd.DataFrame(df11)
print('\nAnswer :')
df11
```

```python
import scipy.stats as stats
# Confidence levels
confidence_levels = [0.90, 0.94, 0.60]
# Calculate and display Z-scores
for confidence_level in confidence_levels:
    alpha = 1 - confidence_level
    # Calculate Z-score for the upper tail (right tail)
    z_score = stats.norm.ppf(1 - alpha / 2)
    print(f"Z-score for a {confidence_level * 100}% confidence interval: {z_score:.3f}")
```

```python
import scipy.stats as stats
sample_size = 25
degree_of_freedom = sample_size - 1
confidence_levels = [0.95, 0.96, 0.99]
# Calculate and display Z-scores
for confidence_level in confidence_levels:
    alpha = 1 - confidence_level
    critical_t_value= stats.t.ppf(1-alpha/2, degree_of_freedom)

    print(f"Critical t-value for a {confidence_level * 100}% confidence interval: {critical_t_value:.3f}")
```

Question 24

```python
from scipy.stats import t
# Given data
sample_mean = 260
population_mean = 270
sample_std = 90
sample_size = 18
# Degrees of freedom
df24 = sample_size - 1
# Calculate the t-score
t_score = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
# Calculate the probability (p-value) using the cumulative distribution function (CDF)
p_value = t.cdf(t_score, df24)
# Probability of having an average life of no more than 260 days
probability = p_value
print("The probability is:", probability)
```
